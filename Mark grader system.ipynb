```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv('student.csv')
data.columns=['gender','group','education','lunch','course','maths','reading score','writing score']
data1=data.groupby('gender')['maths'].agg([np.mean])
data.isnull().sum()
data['total']=data['maths']+data['reading score']+data['writing score']
data['average']=data['total']/3
data['average']=data['average'].astype('int')
data.info
data.groupby(['education','gender'])
data.set_index(['education','gender']).sort_values(['average','total'],ascending=False)
# data['rank']=np.zeros()
def grader(x):
#     for x in data['average']:
        if x>=90 and x<=100:
            print(x)
            return 'A1'
        elif x<90 and x>=80:
            print(x)
            return 'A2'
        elif x>=70 and x<80:
            print(x)
            return 'B'
        elif x>=60 and x<70:
            print(x)
            return 'C'
        elif x>=50 and x<60:
            print(x)
            return 'D'
        else:
            print(x)
            return 'FAIL'

data
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
data.drop(columns=['lunch'],axis=1,inplace=True)
data1=pd.get_dummies(data)
data1['rank']=np.zeros(1000)
data1.isnull().sum()
data1['rank'] = data1['average'].apply(lambda z: grader(z)) 
data1[data1['rank']=='FAIL']
```
